```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

GDP_life_exp = pd.read_csv('all_data.csv')
print(GDP_life_exp.head())

# Has life expectancy increased over time in the six nations?  - line graph(multi?)
sns.lineplot(x='Year', y='Life expectancy at birth (years)', hue='Country', data=GDP_life_exp)
plt.title('Life expectancy over time for each country')
plt.legend(loc="upper right", bbox_to_anchor=(1.5, 1))
plt.show()
plt.clf()

# Has GDP increased over time in the six nations? - line graph (multi?)
sns.lineplot(x='Year', y='GDP', hue='Country', data=GDP_life_exp)
plt.title('GDP over time for each country')
plt.show()
plt.clf()

```

```python
# correlation of GDP and life expectancy of a country? - scatterplot, sub plots

# List of countries in the dataset
countries = GDP_life_exp['Country'].unique()

# Define a list of colors for each country (match the colors to what you want)
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Create a 2x3 grid for the subplots (adjust the number of rows and columns as needed)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Flatten axes array for easier indexing
axes = axes.flatten()

# Loop through countries and create a plot for each
i = 0
for country, color in zip(countries, colors):
    sns.regplot(x='GDP', y='Life expectancy at birth (years)', 
                data=GDP_life_exp[GDP_life_exp['Country'] == country], 
                ax=axes[i], 
                color=color)  # Assign different color to each plot
    axes[i].set_title(f'Country = {country}')
    axes[i].set_xlabel('Life expectancy at birth (years)')
    axes[i].set_ylabel('GDP in Trillions of US Dollars')
    i += 1

# Add a unified legend for all subplots
handles = [plt.Line2D([0], [0], color=color, lw=2) for color in colors]
labels = countries
fig.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.05, 1), ncol=1, title="Country", fontsize=10)

# Adjust layout to avoid overlap
plt.tight_layout(rect=[0, 0, 0.9, 1])  # Adjust the layout to make space on the right
plt.show()


# Display subplots


plt.show()
plt.clf()

```

```python
#What is the average life expectancy in these nations? - box plot?


sns.boxplot(x='Country', y='Life expectancy at birth (years)', data=GDP_life_exp, order=GDP_life_exp["Country"].value_counts(ascending= True).index) 

ax = plt.subplot()
ax.set_xticklabels(GDP_life_exp['Country'].unique(),
rotation=30)

plt.title('Life expectancy of each country')
plt.show()
plt.clf()

```

```python
# What is the distribution of that life expectancy? - histogram 


sns.histplot(GDP_life_exp['Life expectancy at birth (years)'], kde=True)
plt.xlabel('Life expectancy in years')
plt.title('Distribution of Life Expectancy')
plt.ylabel('Frequency')
plt.show()
plt.clf()
```
